```python
n1 = 56
n2 = -345

if n1<0:
    n1 = n1 * (-1)
    
if n2<0:
    n2 = n2 * (-1)


print(f'{n1},{n2}')
Value = n1 - n2
print(f'{Value}=={n1}-{n2}')

if Value <0:
    print(Value)
    Value = Value * (-1)
    print(Value)
else:
    print(Value)
```
