```python
# lets understand the 
#nationality              19
for i in range (null_values['nationality']):

    nullvaue_nationality = main_df[main_df['nationality'].isna()][['country','nationality']].head(1)

    print(f'nullvaue_nationality')
    print(nullvaue_nationality)
    before_update = main_df.iloc[nullvaue_nationality.index][['country','nationality']]
    print(f' -->{before_update}')
    c_name = nullvaue_nationality['country']
    print('--------------')
    print(c_name.values)
    main_df.loc[main_df['nationality'].isna(), 'nationality'] = c_name

    after_update = main_df.iloc[nullvaue_nationality.index][['country','nationality']]
    print(f'after_update -->{after_update}')

null_values= main_df.isna().sum()
#null_values[null_values.values >=1]
```

```python
for i in range (null_values['nationality_full']):
    null_nationality_full= main_df[main_df['nationality_full'].isna()][['country','nationality_full']].sort_values(by='country').head(1)
    null_nationality_full
    temp_index = null_nationality_full.index
    temp_value = null_nationality_full['country']
    print(f'\n temp_value \n {temp_value} \n****************************\n\n\n')

    print(f'\n{main_df.loc[temp_index][['country','nationality_full']]} \n****************************\n\n\n\n ')

    main_df.loc[main_df['nationality_full'].isna(), 'nationality_full'] = temp_value

    print(f'\n{main_df.loc[temp_index][['country','nationality_full']]} \n****************************\n\n\n\n ')

null_values= main_df.isna().sum()
#null_values[null_values.values >=1]
```

```python
#nationality_code        
for i in range (null_values['nationality_code']):
    null_values_nationality_code= main_df[main_df['nationality_code'].isna()][['country','nationality_code']].sort_values(by='country').head(1)
    temp_index =null_values_nationality_code.index
    temp_value =null_values_nationality_code['country']
    print(f'\n null_values_nationality_code \n{temp_index}\n ===========')
    print(f'\n null_values_nationality_code \n{temp_value}\n ===========')

    print(f'===before update ===\n {main_df.loc[temp_index][['country','nationality_code']]}')
    main_df.loc[main_df['nationality_code'].isna(), 'nationality_code'] = temp_value


    print(f'===after update ===\n {main_df.loc[temp_index][['country','nationality_code']]}')

null_values= main_df.isna().sum()
#null_values[null_values.values >=1]
```
