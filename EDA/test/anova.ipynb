```python
#low_range = int(input("Enter the lower range: "))
#high_range = int(input("Enter the higher range: "))
#arm_values = []
#low_range = 1000
#high_range = 2000
#arm_values = []
low_range = 1000
high_range = 2000
arm_values = []

for number in range(152, 155): #152,153,154
    
    arm = 0
    temp = number          
    print(f'temp--> {temp}, number --> {number}')
    n = len(str(number))
    print(f'{number}---> n length --> {n}')  
    while temp > 0:
        print('==while loop starts===== ')
        rem = temp % 10
        print(f'{rem} = {temp} % 10 ')
        arm += rem ** n
        print(f'{arm} += {rem} ** {n}')
        temp //= 10
        print(f'{temp } temp //= 10')

    if arm == number:
        print('==if loop starts===== ')
        arm_values.append(number)

print("Armstrong numbers:", arm_values)
```
