```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('mtcars.csv')
```

```python
line_break = "=================================================================================================================================================================="
```

```python
df.isna().sum()
```

```python
print(df.columns)
```

```python
mean_valeu =df[df['wt']>0]['wt'].mean()

mean_valeu

```

```python
df['wt'].fillna(3, inplace=True)
```

```python
plt.boxplot(df['wt'])
plt.show()
```

```python
desc = df['wt'].describe()

print(desc)
print(line_break)

Q1 = desc['25%']
Q3 = desc['75%']


IQR = Q3-Q1

print(f'q1==> {Q1}, q3 --> {Q3}')

print(f'iqr=={IQR}')

print(line_break)
LOWER_BOUND = Q1-1.5 * IQR
UPPER_BOUND = Q3 + 1.5 * IQR

print(f'LOWER_BOUND --> {LOWER_BOUND}')

print(f'UPPER_BOUND --> {UPPER_BOUND}')
print(line_break)
outliers = df[(df['wt'] < LOWER_BOUND) | (df['wt'] > UPPER_BOUND)]
#outliers
print(line_break)
print(outliers)
```

```python
for i in (outliers.index):
    print(i)
    
    if df.loc[i,'wt'] >= UPPER_BOUND:
        df.loc[i,'wt'] = 5
        print('upper')
        print(df.loc[i,'wt'])
    else:
        df.loc[i,'wt'] = 1
        print(df.loc[i,'wt'])
        print('lower')
```

```python
desc = df['wt'].describe()

print(desc)
print(line_break)

Q1 = desc['25%']
Q3 = desc['75%']


IQR = Q3-Q1

print(f'q1==> {Q1}, q3 --> {Q3}')

print(f'iqr=={IQR}')

print(line_break)
LOWER_BOUND = Q1-1.5 * IQR
UPPER_BOUND = Q3 + 1.5 * IQR

print(f'LOWER_BOUND --> {LOWER_BOUND}')

print(f'UPPER_BOUND --> {UPPER_BOUND}')
print(line_break)
outliers = df[(df['wt'] < LOWER_BOUND) | (df['wt'] > UPPER_BOUND)]
#outliers
print(line_break)
print(outliers)
```
