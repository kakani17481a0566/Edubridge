That is, the `plot()` method on pandas’ Series and DataFrame is a wrapper around `plt.plot()` we'll see in upcoming lectures.

Pandas can easily read data stored in different file formats like CSV, JSON, XML or even Excel.

With CSV files all you need is a single line to load in the data:

```python
import pandas as pd
```

> We will be working with [**Amazon's Top 50 bestselling books from 2009 to 2019**](https://www.kaggle.com/sootersaalu/amazon-top-50-bestselling-books-2009-2019?select=bestsellers+with+categories.csv) dataset.

```python
df = pd.read_csv('data/bestsellers.csv')

df
```

<h1 style="color: white; font-weight: bold;">
    Plotting basics
</h1>

<br><br>

Let's see some common used type of plots:

```python
df.plot?
```

## Scatter plot
![orange-divider](https://user-images.githubusercontent.com/7065401/98619088-44ab6000-22e1-11eb-8f6d-5532e68ab274.png)


```python
df.plot(kind='scatter', x='Reviews', y='Price')
```

```python
df.plot(kind='scatter',
        x='Reviews',
        y='Price',
        color='orange',
        title='Reviews vs. Price',
        figsize=(12, 6))
```

## Bar plot
![orange-divider](https://user-images.githubusercontent.com/7065401/98619088-44ab6000-22e1-11eb-8f6d-5532e68ab274.png)


We can also make some nice **bar plots** with pandas.

Let's see the average reviews count per year:

```python
df_year = df.groupby('Year').mean().reset_index()
df_year
```

```python
df_year.plot(kind='bar',
             x='Year',
             y='Reviews')
```

```python
df_year.plot(kind='bar',
             x='Year',
             y='Reviews',
             color='lightgreen',
             title='Reviews per Year',
             figsize=(12, 6))
```

Also is possible with columns without numeric values:

```python
top_authors = df['Author'].value_counts().head(10)
top_authors
```

```python
top_authors.plot(kind='bar',
                 color='lightblue',
                 title='Top book quantity per Author',
                 figsize=(12, 6))
```

## Line plot
![orange-divider](https://user-images.githubusercontent.com/7065401/98619088-44ab6000-22e1-11eb-8f6d-5532e68ab274.png)

Another useful type of plot is the **line plot**.

```python
df_year['User Rating']
```

```python
df_year['User Rating'].plot(kind='line')
```

```python
df_year['User Rating'].plot(color='purple',
                            linewidth=4,
                            title='Average User Rating per Year',
                            figsize=(12, 6))
```

```python
df_year['Reviews'].plot(color='salmon',
                        linewidth=4,
                        title='Average Reviews per Year',
                        figsize=(12, 6))
```

## Pie plot
![orange-divider](https://user-images.githubusercontent.com/7065401/98619088-44ab6000-22e1-11eb-8f6d-5532e68ab274.png)

In order to see the fractional area of different values within a column, we can make a **pie plot**.

```python
df_year
```

```python
df_year['Reviews'].plot(kind='pie')
```

```python
df_year['Reviews'].plot(kind='pie',
                        labels=df_year['Year'],
                        title='Reviews per Year',
                        figsize=(8,8))
```

## Histograms
![orange-divider](https://user-images.githubusercontent.com/7065401/98619088-44ab6000-22e1-11eb-8f6d-5532e68ab274.png)

Histograms can also be drawn.

```python
df
```

```python
df['Price'].plot(kind='hist')
```

```python
df['Price'].plot(kind='hist',
                 rwidth=0.9,
                 color="pink",
                 title="Price histogram",
                 figsize=(12, 6))
```

```python
df['Price'].plot(kind='hist',
                 rwidth=0.9,
                 orientation='horizontal',
                 color="pink",
                 title="Price histogram",
                 figsize=(12, 6))
```

## KDE plot
![orange-divider](https://user-images.githubusercontent.com/7065401/98619088-44ab6000-22e1-11eb-8f6d-5532e68ab274.png)

We can also generate Kernel Density Estimate plot (KDE) using Gaussian kernels.

In statistics, kernel density estimation (KDE) is a non-parametric way to estimate the probability density function (PDF) of a random variable. This function uses Gaussian kernels and includes automatic bandwidth determination.

```python
df['Price'].plot(kind='kde')
```

## Box plot
![orange-divider](https://user-images.githubusercontent.com/7065401/98619088-44ab6000-22e1-11eb-8f6d-5532e68ab274.png)


```python
df['Reviews'].plot(kind='box')
```

```python
df[['Reviews', 'User Rating']].plot(kind='box')
```

```python
df[['Reviews', 'User Rating']].plot(kind='box',
                                    subplots=True,
                                    layout=(1,2))
```

```python
df[['Reviews', 'User Rating']].plot(kind='box',
                                    subplots=True,
                                    layout=(2,1))
```

Plots can be saved into images of multiple formats with the `plt.savefig` function:

```python
plot = df.plot(kind='scatter', x='Reviews', y='Price')
```

```python
plot.figure.savefig('plot1.png')
```
