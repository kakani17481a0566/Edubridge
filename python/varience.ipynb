```python
nums = [23, 21, 26, 32, 15, 28, 30, 38, 31, 25]
sum = 0
len_nums = len(nums)
for i in nums:
    sum += i

print(f"sum--> {sum}")
print(f"len--> {len_nums}")

avg_num = sum / len_nums


print(f"avg--> {avg_num}")

diff_sqare = []

for i in nums:
    value = (i - avg_num) ** 2
    diff_sqare.append(value)

print(f'diff_sqare --> {diff_sqare}')



sum_sqare = 0
for i in diff_sqare:
    sum_sqare += i

print(f'sum_sqare --> sum_sqare')

variance = sum_sqare / len_nums
print(f'varience --> {variance}')
print(f'The standed dev ---> {variance**0.5}')
```

```python
import matplotlib.pyplot as plt

# Data to plot
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
explode = (0.1, 0, 0, 0)  # explode 1st slice

# Plot
plt.figure(figsize=(7, 7))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title('Pie Chart Example')
plt.show()
```

labels: A list of labels for each slice of the pie.
sizes: A list of sizes corresponding to each slice.
colors: Optional list of colors for each slice.
explode: A tuple indicating how much each slice should be offset from the center.
autopct: Format string to label the slices with their percentage values.
shadow: Whether to draw a shadow beneath the pie.
startangle: Angle by which the pie chart is rotated.
